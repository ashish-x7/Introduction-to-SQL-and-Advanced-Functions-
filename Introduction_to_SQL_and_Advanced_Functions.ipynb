### Question 1: Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

### Answer 1: Understanding DDL, DML, and DQL in SQL

SQL (Structured Query Language) commands are categorized into three main types based on their function:

1.  **DDL (Data Definition Language)**
    *   **Purpose:** DDL commands are used to *define*, *modify*, or *delete* the structure of database objects (like tables, indexes, views, etc.). They deal with the schema of the database.
    *   **Focus:** Managing the database's blueprint.
    *   **Effect:** Changes made by DDL commands are permanent and automatically committed (saved) to the database.
    *   **Examples:** `CREATE`, `ALTER`, `DROP`, `TRUNCATE`, `RENAME`.
    *   **Example:**
        ```sql
        CREATE TABLE Employees (
            EmployeeID INT PRIMARY KEY,
            FirstName VARCHAR(50),
            LastName VARCHAR(50),
            HireDate DATE
        );
        ```

2.  **DML (Data Manipulation Language)**
    *   **Purpose:** DML commands are used to *manage* data within the database objects. They allow you to add, retrieve, modify, and delete data rows.
    *   **Focus:** Managing the actual data content.
    *   **Effect:** Changes made by DML commands are *not* automatically committed. They require a `COMMIT` statement to save changes or a `ROLLBACK` to undo them.
    *   **Examples:** `INSERT`, `UPDATE`, `DELETE`.
    *   **Example:**
        ```sql
        INSERT INTO Employees (EmployeeID, FirstName, LastName, HireDate)
        VALUES (1, 'John', 'Doe', '2023-01-15');
        ```

3.  **DQL (Data Query Language)**
    *   **Purpose:** DQL commands are used to *retrieve* data from the database. While sometimes grouped under DML because it's a form of data manipulation (reading), it's often distinguished because it doesn't modify the data.
    *   **Focus:** Retrieving specific data sets.
    *   **Effect:** DQL commands do not change the database state; they simply fetch information.
    *   **Examples:** `SELECT`.
    *   **Example:**
        ```sql
        SELECT FirstName, LastName FROM Employees WHERE EmployeeID = 1;
        ```

**In essence:**
*   **DDL** is for building and modifying the *house* (database structure).
*   **DML** is for furnishing and maintaining the *contents* of the house (data).
*   **DQL** is for looking at specific rooms or items *in* the house (querying data).

### Question 2: What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

### Answer 2: The Purpose and Types of SQL Constraints

**Purpose of SQL Constraints:**
SQL constraints are rules enforced on data columns in a table. Their primary purpose is to **limit the type of data that can go into a table**, thereby ensuring the **accuracy and reliability of the data**. Constraints are used to:

*   **Enforce Data Integrity:** Prevent the entry of invalid, inconsistent, or duplicate data.
*   **Maintain Relationships:** Ensure that links between tables (e.g., parent-child relationships) are correctly maintained.
*   **Improve Database Performance:** Some constraints, like PRIMARY KEYs, automatically create indexes which can speed up data retrieval.


**Three Common Types of Constraints:**

1.  **PRIMARY KEY Constraint**
    *   **Description:** A PRIMARY KEY uniquely identifies each record in a table. It must contain unique values, and it cannot contain NULL values. Each table can have only one primary key, which can consist of one or more columns.
    *   **Scenario:** Imagine a `Students` table. Each student needs a unique identifier. An `StudentID` column with a PRIMARY KEY constraint would ensure that no two students have the same ID and that every student record has an ID.
        ```sql
        CREATE TABLE Students (
            StudentID INT PRIMARY KEY,
            FirstName VARCHAR(50),
            LastName VARCHAR(50)
        );
        ```

2.  **FOREIGN KEY Constraint**
    *   **Description:** A FOREIGN KEY is a field (or collection of fields) in one table that refers to the PRIMARY KEY in another table. It establishes and enforces a link between two tables, ensuring referential integrity. This means you cannot insert a row into the child table if the referenced value doesn't exist in the parent table's primary key.
    *   **Scenario:** Consider an `Orders` table and a `Customers` table. Each order must be placed by an existing customer. A `CustomerID` column in the `Orders` table, defined as a FOREIGN KEY referencing the `CustomerID` (PRIMARY KEY) in the `Customers` table, would prevent orders from being created for non-existent customers.
        ```sql
        CREATE TABLE Customers (
            CustomerID INT PRIMARY KEY,
            CustomerName VARCHAR(100)
        );

        CREATE TABLE Orders (
            OrderID INT PRIMARY KEY,
            OrderDate DATE,
            CustomerID INT,
            FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
        );
        ```

3.  **NOT NULL Constraint**
    *   **Description:** The NOT NULL constraint ensures that a column cannot have a NULL value. This means that you must provide a value for that column when inserting or updating a record.
    *   **Scenario:** In an `Employees` table, it's essential that every employee has a `FirstName` and `LastName`. Applying a NOT NULL constraint to these columns would prevent records from being inserted or updated without these critical pieces of information.
        ```sql
        CREATE TABLE Employees (
            EmployeeID INT PRIMARY KEY,
            FirstName VARCHAR(50) NOT NULL,
            LastName VARCHAR(50) NOT NULL,
            Email VARCHAR(100)
        );
        ```

### Question 3: Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

### Answer 3: Understanding LIMIT and OFFSET for Pagination

Both `LIMIT` and `OFFSET` clauses are used in SQL to control which rows are returned by a `SELECT` statement, typically for pagination or to retrieve a specific subset of data.

1.  **`LIMIT` Clause**
    *   **Purpose:** The `LIMIT` clause is used to restrict the number of rows returned by a query. It specifies the maximum number of rows that the result set should contain.
    *   **Syntax:** `SELECT column1, column2 FROM table_name LIMIT count;`
    *   **Example:** `SELECT * FROM Products LIMIT 5;` (This would return the first 5 products from the table.)

2.  **`OFFSET` Clause**
    *   **Purpose:** The `OFFSET` clause is used to skip a specified number of rows before returning the result set. It's often used in conjunction with `LIMIT` to implement pagination, allowing you to retrieve subsequent "pages" of data.
    *   **Syntax:** `SELECT column1, column2 FROM table_name OFFSET start;`
    *   **Example:** `SELECT * FROM Products OFFSET 10;` (This would skip the first 10 products and return all products starting from the 11th one.)

### Using LIMIT and OFFSET for Pagination (Retrieving the Third Page)

To retrieve a specific page of results, you combine `LIMIT` and `OFFSET`. The `LIMIT` specifies the number of records per page, and the `OFFSET` calculates how many records to skip based on the page number.

**Scenario:** Retrieve the third page of results, with each page having 10 records.

*   **Page Size (LIMIT):** 10 records
*   **Desired Page:** 3rd page

To get the 3rd page, you need to skip the records from the 1st page (10 records) and the 2nd page (another 10 records). So, you need to skip a total of 20 records.

**Calculation for OFFSET:** `(Page Number - 1) * Records Per Page`
For the 3rd page: `(3 - 1) * 10 = 2 * 10 = 20`

**SQL Query:**
```sql
SELECT *
FROM YourTable
ORDER BY SomeColumn -- Important for consistent pagination
LIMIT 10
OFFSET 20;
```

**Explanation of the query:**
*   `SELECT * FROM YourTable`: Selects all columns from `YourTable`.
*   `ORDER BY SomeColumn`: **Crucially important!** Always use an `ORDER BY` clause when paginating. Without it, the order of results is not guaranteed, and subsequent pages might return inconsistent or duplicate data. You should order by a column (or set of columns) that provides a consistent and deterministic sort order.
*   `LIMIT 10`: Specifies that we want to retrieve a maximum of 10 records.
*   `OFFSET 20`: Specifies that we want to skip the first 20 records before starting to retrieve the 10 records for the current page.

This combination effectively fetches the 10 records that constitute the third page of results.

### Question 4: What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage

### Answer 4: Understanding Common Table Expressions (CTEs)

A **Common Table Expression (CTE)** is a temporary, named result set that you can reference within a single `SELECT`, `INSERT`, `UPDATE`, or `DELETE` statement. It's essentially a named subquery that you can define once and reuse multiple times within a more complex query. CTEs are defined using the `WITH` clause.

**Main Benefits of CTEs:**

1.  **Improved Readability and Organization:** Complex queries can be broken down into smaller, logical, and more manageable building blocks. This makes the SQL code easier to understand, debug, and maintain.
2.  **Enhanced Reusability:** A CTE can be referenced multiple times within the same query without having to rewrite the same subquery logic repeatedly. This reduces redundancy.
3.  **Simplifies Complex Logic:** For recursive queries or scenarios where you need to perform multi-step data transformations (e.g., filtering, aggregating, then joining), CTEs provide a clear and sequential way to structure these steps.
4.  **Better Performance (in some cases):** While not always guaranteed to improve performance over subqueries, CTEs can sometimes lead to more optimized query plans because the database engine can better understand and optimize the separate, named steps.
5.  **Recursive Queries:** CTEs are essential for writing recursive queries, which are queries that refer to themselves, useful for traversing hierarchical or graph-like data (e.g., organizational charts, bill of materials).


**Simple SQL Example Demonstrating CTE Usage:**

Let's imagine you have a table called `Sales` with `ProductID`, `SaleAmount`, and `SaleDate`. You want to find the total sales for each product, but only for products whose total sales exceed a certain threshold (e.g., $1000).

Without a CTE, you might write a subquery:
```sql
SELECT ProductID, TotalSales
FROM (
    SELECT ProductID, SUM(SaleAmount) AS TotalSales
    FROM Sales
    GROUP BY ProductID
) AS ProductSalesSummary
WHERE TotalSales > 1000;
```

Using a CTE, the same logic becomes much clearer:
```sql
WITH ProductSalesSummary AS (
    SELECT
        ProductID,
        SUM(SaleAmount) AS TotalSales
    FROM
        Sales
    GROUP BY
        ProductID
)
SELECT
    ProductID,
    TotalSales
FROM
    ProductSalesSummary
WHERE
    TotalSales > 1000;
```

In this example:
*   The `WITH ProductSalesSummary AS (...)` block defines our CTE, giving it the name `ProductSalesSummary`.
*   Inside the CTE, we calculate the `TotalSales` for each `ProductID`.
*   The main `SELECT` statement then treats `ProductSalesSummary` as if it were a regular table, allowing us to easily filter its results based on `TotalSales > 1000`.

This structure makes it very easy to follow the logic: first, summarize sales by product, then filter those summarized results.

### Question 5: Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

### Answer 5: SQL Normalization and the First Three Normal Forms

**Concept of SQL Normalization:**
SQL Normalization is a systematic process of organizing the columns and tables of a relational database to minimize data redundancy and improve data integrity. It involves decomposing a large table into smaller, related tables and defining relationships between them. This process helps to ensure that data is stored logically and efficiently.

**Primary Goals of Normalization:**

1.  **Eliminate Data Redundancy:** Avoid storing the same data multiple times in the database, which saves storage space and prevents inconsistencies.
2.  **Ensure Data Integrity:** Maintain the accuracy and consistency of data. When data is stored in only one place, there's less chance of errors or discrepancies.
3.  **Improve Data Efficiency:** Make the database easier to maintain and update. Changes to data only need to be made in one place.
4.  **Enhance Query Performance (indirectly):** While complex joins might be needed, a well-normalized database can lead to more efficient queries by retrieving only necessary data and reducing the amount of data processed.
5.  **Simplify Database Structure:** Make the database schema more logical and easier to understand.

**Brief Explanation of the First Three Normal Forms:**

Database normalization is typically carried out in stages, with each stage representing a "normal form." The most common normal forms are 1NF, 2NF, and 3NF.

1.  **First Normal Form (1NF):**
    *   **Rule:** A table is in 1NF if it meets two basic criteria:
        *   **Atomic Values:** Each column must contain atomic (indivisible) values, meaning there are no repeating groups of columns or multi-valued attributes within a single cell.
        *   **Unique Rows:** Each row must be unique, meaning there's a primary key that uniquely identifies each record.
    *   **Example (Not 1NF):** A table `Customers` with a column `PhoneNumbers` containing "123-4567, 987-6543" in a single cell. This violates atomicity.
    *   **To achieve 1NF:** Separate the multi-valued attribute into separate rows or a separate table.

2.  **Second Normal Form (2NF):**
    *   **Rule:** A table is in 2NF if it is already in 1NF **AND** all non-key attributes are fully functionally dependent on the entire primary key. This means that if a primary key is composite (made of two or more columns), no non-key attribute should depend on only *part* of the primary key.
    *   **Example (Not 2NF):** Consider a table `Order_Details` with a composite primary key `(OrderID, ProductID)` and non-key attributes `ProductName` and `Quantity`. If `ProductName` depends only on `ProductID` (part of the primary key) and not on `OrderID`, then it's not in 2NF.
    *   **To achieve 2NF:** Remove attributes that depend on only part of a composite primary key and place them in a separate table, linking them with a foreign key.

3.  **Third Normal Form (3NF):**
    *   **Rule:** A table is in 3NF if it is already in 2NF **AND** there are no transitive dependencies. A transitive dependency occurs when a non-key attribute is dependent on another non-key attribute.
    *   **Example (Not 3NF):** In an `Employees` table, if `EmployeeID` is the primary key, `DepartmentID` is a non-key attribute, and `DepartmentName` is another non-key attribute that depends on `DepartmentID` (i.e., `EmployeeID` determines `DepartmentID`, and `DepartmentID` determines `DepartmentName`), then there's a transitive dependency `EmployeeID -> DepartmentID -> DepartmentName`.
    *   **To achieve 3NF:** Remove attributes that are transitively dependent on the primary key and place them in a separate table.

### Question 7: Generate a report showing CustomerName, Email, and the TotalNumberofOrders for each customer. Include customers who have not placed any orders, in which case their TotalNumberofOrders should be 0. Order the results by CustomerName.

### Answer 7: Customer Order Report

To achieve this report, we'll perform the following steps:
1.  Select `CustomerName` and `Email` from the `Customers` table.
2.  Use a `LEFT JOIN` with the `Orders` table to include all customers.
3.  Count the `OrderID`s for each customer using `COUNT(Orders.OrderID)`.
4.  Employ `IFNULL(COUNT(Orders.OrderID), 0)` to ensure that customers without orders show a count of 0 instead of `NULL`.
5.  Group the results by customer details and order them by `CustomerName`.

````python
report_sql = """
SELECT
    c.CustomerName,
    c.Email,
    IFNULL(COUNT(o.OrderID), 0) AS TotalNumberofOrders
FROM
    Customers AS c
LEFT JOIN
    Orders AS o ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID, c.CustomerName, c.Email
ORDER BY
    c.CustomerName;
"""

cursor.execute(report_sql)
report_results = cursor.fetchall()

print("Customer Order Report:")
print("---------------------------------------------------")
print(f"{'CustomerName':<20} {'Email':<25} {'TotalOrders':<10}")
print("---------------------------------------------------")
for row in report_results:
    print(f"{row[0]:<20} {row[1]:<25} {row[2]:<10}")
print("---------------------------------------------------")
````

### Question 8: Retrieve Product Information with Category: Write a SQL query to display the ProductName, Price, StockQuantity, and CategoryName for all products. Order the results by CategoryName and then ProductName alphabetically.

To retrieve the requested product information, we will:

### Question 9: Write a SQL query that uses a Common Table Expression (CTE) and a Window Function (specifically ROW_NUMBER() or RANK()) to display the CategoryName, ProductName, and Price for the top 2 most expensive products in each CategoryName.

### Answer 9: Top 2 Most Expensive Products per Category using CTE and Window Function

To find the top 2 most expensive products in each category, we will:
1.  **Define a Common Table Expression (CTE)** named `RankedProducts`.
2.  Inside the CTE, `SELECT` the `CategoryName`, `ProductName`, and `Price`.
3.  Use the `ROW_NUMBER()` window function, `PARTITION`ing by `CategoryName` and `ORDER BY` `Price` in descending order. This assigns a rank to each product within its category based on price.
4.  `JOIN` the `Products` and `Categories` tables to get the category names.
5.  Finally, `SELECT` from the `RankedProducts` CTE where the `rn` (row number) is less than or equal to 2, effectively fetching the top 2 products per category.
6.  Order the final results by `CategoryName` and then `Price` in descending order for clarity.

````python
top_products_sql = """
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC) as rn
    FROM
        Products AS p
    JOIN
        Categories AS c ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM
    RankedProducts
WHERE
    rn <= 2
ORDER BY
    CategoryName, Price DESC;
"""

cursor.execute(top_products_sql)
top_products_results = cursor.fetchall()

print("Top 2 Most Expensive Products per Category:")
print("-----------------------------------------------------------")
print(f"{'CategoryName':<20} {'ProductName':<20} {'Price':<10}")
print("-----------------------------------------------------------")
for row in top_products_results:
    print(f"{row[0]:<20} {row[1]:<20} {row[2]:<10.2f}")
print("-----------------------------------------------------------")
````

## Analyze Top 5 Customers by Amount Spent

### Subtask:
Write a SQL query to identify the top 5 customers based on their total spending from the mock Sakila database. The report should include the customer's name, email, and the aggregated total amount spent. Display the results.


**Reasoning**:
To identify the top 5 customers by total spending, I need to write a SQL query that joins the `customer` and `payment` tables, aggregates the `amount`, orders by `total_spent` in descending order, and limits the result to 5. This query will then be executed, and the results will be fetched and displayed.



````python
top_customers_sql = '''
SELECT
    c.first_name || ' ' || c.last_name AS CustomerName,
    c.email AS Email,
    SUM(p.amount) AS TotalSpent
FROM
    customer AS c
JOIN
    payment AS p ON c.customer_id = p.customer_id
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
ORDER BY
    TotalSpent DESC
LIMIT 5;
'''

cursor.execute(top_customers_sql)
top_customers_results = cursor.fetchall()

print("\nTop 5 Customers by Total Spending:")
print("------------------------------------------------------------------")
print(f"{'Customer Name':<25} {'Email':<30} {'Total Spent':<15}")
print("------------------------------------------------------------------")
for row in top_customers_results:
    print(f"{row[0]:<25} {row[1]:<30} {row[2]:<15.2f}")
print("------------------------------------------------------------------")
````

## Analyze Top 3 Movie Categories by Rental Count

### Subtask:
Determine which 3 movie categories have the highest rental counts in the mock Sakila database. The output should show the category name and the total number of times movies from that category were rented. Display the results.


**Reasoning**:
To achieve the subtask of finding the top 3 movie categories by rental count, I need to construct a SQL query that joins the relevant tables, aggregates the rental counts per category, orders them, and limits the result set. After executing the query, the results will be fetched and printed in a formatted manner.



````python
top_categories_sql = '''
SELECT
    c.name AS CategoryName,
    COUNT(r.rental_id) AS RentalCount
FROM
    category AS c
JOIN
    film_category AS fc ON c.category_id = fc.category_id
JOIN
    film AS f ON fc.film_id = f.film_id
JOIN
    inventory AS i ON f.film_id = i.film_id
JOIN
    rental AS r ON i.inventory_id = r.inventory_id
GROUP BY
    c.name
ORDER BY
    RentalCount DESC
LIMIT 3;
'''

cursor.execute(top_categories_sql)
top_categories_results = cursor.fetchall()

print("\nTop 3 Movie Categories by Rental Count:")
print("---------------------------------------------------")
print(f"{'Category Name':<20} {'Rental Count':<15}")
print("---------------------------------------------------")
for row in top_categories_results:
    print(f"{row[0]:<20} {row[1]:<15}")
print("---------------------------------------------------")
````

## Calculate Film Availability and Unrented Films per Store

### Subtask:
Develop a SQL query to calculate, for each store in the mock Sakila database, the total number of films available and how many of those films have never been rented. Display the results.


**Reasoning**:
To address the subtask, I need to construct a SQL query that calculates the total number of films available per store and the count of unrented films for each store. This involves joining the `store`, `inventory`, and `rental` tables, performing conditional aggregation, and then formatting and displaying the results.



````python
film_availability_sql = '''
SELECT
    s.store_id,
    COUNT(DISTINCT i.film_id) AS TotalFilmsAvailable,
    COUNT(CASE WHEN r.rental_id IS NULL THEN i.inventory_id ELSE NULL END) AS UnrentedFilms
FROM
    store AS s
LEFT JOIN
    inventory AS i ON s.store_id = i.store_id
LEFT JOIN
    rental AS r ON i.inventory_id = r.inventory_id
GROUP BY
    s.store_id
ORDER BY
    s.store_id;
'''

cursor.execute(film_availability_sql)
film_availability_results = cursor.fetchall()

print("\nFilm Availability and Unrented Films per Store:")
print("---------------------------------------------------")
print(f"{'Store ID':<10} {'Total Films Available':<25} {'Unrented Films':<20}")
print("---------------------------------------------------")
for row in film_availability_results:
    print(f"{row[0]:<10} {row[1]:<25} {row[2]:<20}")
print("---------------------------------------------------")
````

## Analyze Monthly Revenue for 2023

### Subtask:
Construct a SQL query to show the total revenue per month for the year 2023 from the mock Sakila database. This will help analyze business seasonality. Display the results.


**Reasoning**:
To analyze monthly revenue for 2023, I need to construct a SQL query that extracts the month from the payment date, sums the payment amounts, filters for the year 2023, groups by month, and orders the results chronologically. After executing the query, I will fetch and display the results.



````python
monthly_revenue_sql = '''
SELECT
    STRFTIME('%Y-%m', payment_date) AS PaymentMonth,
    SUM(amount) AS TotalRevenue
FROM
    payment
WHERE
    STRFTIME('%Y', payment_date) = '2023'
GROUP BY
    PaymentMonth
ORDER BY
    PaymentMonth;
'''

cursor.execute(monthly_revenue_sql)
monthly_revenue_results = cursor.fetchall()

print("\nMonthly Revenue for 2023:")
print("-----------------------------------")
print(f"{'Month':<15} {'Total Revenue':<15}")
print("-----------------------------------")
for row in monthly_revenue_results:
    print(f"{row[0]:<15} {row[1]:<15.2f}")
print("-----------------------------------")
````

## Identify Frequent Renters (last 6 months)

### Subtask:
Write a SQL query to identify customers who have rented more than 10 times within the last 6 months in the mock Sakila database. Display the customer details.


**Reasoning**:
To identify frequent renters within the last 6 months, I need to construct a SQL query that joins the `customer` and `rental` tables. I will first determine the latest rental date to establish a relative 'last 6 months' window, then filter rentals within that period, group by customer, count their rentals, and filter for those with more than 1 rental, finally displaying the results.



````python
import datetime

# 1. Get the maximum rental_date from the rental table to establish 'last 6 months' relative to the data
cursor.execute("SELECT MAX(rental_date) FROM rental;")
max_rental_date_str = cursor.fetchone()[0]

# If no rental_date exists, handle it or set a default
if max_rental_date_str:
    max_rental_date = datetime.datetime.strptime(max_rental_date_str, '%Y-%m-%d %H:%M:%S')
    six_months_ago = max_rental_date - datetime.timedelta(days=6*30) # Approximate 6 months
    six_months_ago_str = six_months_ago.strftime('%Y-%m-%d %H:%M:%S')
else:
    six_months_ago_str = '1900-01-01 00:00:00' # A very old date if no rentals exist

frequent_renters_sql = f'''
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email,
    COUNT(r.rental_id) AS rental_count
FROM
    customer AS c
JOIN
    rental AS r ON c.customer_id = r.customer_id
WHERE
    r.rental_date >= '{six_months_ago_str}'
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
HAVING
    COUNT(r.rental_id) > 1
ORDER BY
    rental_count DESC;
'''

cursor.execute(frequent_renters_sql)
frequent_renters_results = cursor.fetchall()

print("\nFrequent Renters (more than 1 rental in last 6 months):")
print("-------------------------------------------------------------------------------------------------")
print(f"{'Customer ID':<12} {'Customer Name':<25} {'Email':<30} {'Rentals':<10}")
print("-------------------------------------------------------------------------------------------------")
for row in frequent_renters_results:
    customer_name = f"{row[1]} {row[2]}"
    print(f"{row[0]:<12} {customer_name:<25} {row[3]:<30} {row[4]:<10}")
print("-------------------------------------------------------------------------------------------------")
````
